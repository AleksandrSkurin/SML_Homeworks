```r
set.seed(222)
len = length(Auto$mpg)
med = median(Auto$mpg)
mpg = rep(NA,len) 

for (i in 1:len) if (Auto$mpg[i] > med) mpg[i]=1 else mpg[i]=0

df = Auto
df$mpg = as.factor(mpg)
```

```r
# b)

lin=tune(svm,mpg~.,data=df,kernel='linear',
                 ranges=list(cost=c(0.001,0.01,0.1,1,5,10,100,1000)))
summary(lin)
```

```r
# c)

set.seed(222)

rad=tune(svm, mpg~., data=df, kernel='radial',
                 ranges=list(cost=c(0.1,1,10,100,1000),gamma=c(0.5,1,2,3,4)))
rad$best.parameters
rad$best.performance
```
