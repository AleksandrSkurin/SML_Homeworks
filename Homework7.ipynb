```r
# Q1

set.seed(1)

data <- read.csv('C:/Users/ashkurin/Documents/SML/SML_Section7/data/bankruptcy_prediction/data.csv')

split <- sample(c(rep(0, 0.8 * nrow(data)), 
                        rep(1, 0.2 * nrow(data))))

table(split)

train <- data[split == 0, ]

test <- data[split == 1, ]
```

```r
str(train)
```

```r
# Q2)

install.packages('neuralnet')
library(neuralnet)
```

```r
nn=neuralnet(Bankrupt.~.,data=train, hidden=1,act.fct = "logistic",linear.output = FALSE)
plot(nn)
```

```r
Predict=compute(nn,test)
Predict$net.result
```
